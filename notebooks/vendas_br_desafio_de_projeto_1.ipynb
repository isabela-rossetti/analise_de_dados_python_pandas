#**Trabalhando com Planilhas do Excel**

```python
#Importando as bibliotecas necessárias para a análise
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Importando os dados do GitHub
df1 = pd.read_excel('https://github.com/isabela-rossetti/analise_de_dados_python_pandas/blob/main/datasets/Aracaju.xlsx?raw=true')
df2 = pd.read_excel('https://github.com/isabela-rossetti/analise_de_dados_python_pandas/blob/main/datasets/Fortaleza.xlsx?raw=true')
df3 = pd.read_excel('https://github.com/isabela-rossetti/analise_de_dados_python_pandas/blob/main/datasets/Natal.xlsx?raw=true')
df4 = pd.read_excel('https://github.com/isabela-rossetti/analise_de_dados_python_pandas/blob/main/datasets/Recife.xlsx?raw=true')
df5 = pd.read_excel('https://github.com/isabela-rossetti/analise_de_dados_python_pandas/blob/main/datasets/Salvador.xlsx?raw=true')
```

```python
df2.head()
```

```python
#Conferindo tipos de dados, tamanho do df e valores nulos
print(df1.info())
print(df2.info())
print(df3.info())
print(df4.info())
print(df5.info())
```

```python
#juntando todos os arquivos
df = pd.concat([df1,df2,df3,df4,df5])
```

```python
#Exibindo as 5 primeiras linhas
df.head()
```

```python
#Exibindo as 5 últimas linhas
df.tail()
```

```python
df.sample(5)
```

```python
#Verificando o tipo de dado de cada coluna, tamanho do df e valores nulos
df.info()
```

```python
#Alterando o tipo da variável da coluna LojaID
df['LojaID'] = df['LojaID'].astype('object')
```

```python
df.info()
```

```python
df.head()
```

**Criando colunas novas**

```python
#Criando a coluna de receita
df['Receita'] = df['Vendas'].mul(df['Qtde'])
```

```python
df.head()
```

```python
#Criando coluna para calcular quantidade (caso não houvesse essa coluna do df)
df['Receita/Vendas'] = df['Receita'] / df['Vendas'] 
```

```python
df.head()
```

```python
#Retornando a maior receita
df['Receita'].max()
```

```python
#Retornando a menor receita
df['Receita'].min()
```

```python
#Retornando as 10 maiores receitas
df.nlargest(10, 'Receita')
```

```python
#Retornando as 10 menores receitas
df.nsmallest(10, 'Receita')
```

```python
#Agrupamento de venda por Cidade e LojaID
df.groupby(['Cidade', 'LojaID'])['Receita'].mean().round(2)
```

```python
#Agrupamento por cidade
df.groupby('Cidade')['Receita'].sum()
```

```python
#Ordenando o conjunto de dados
df.sort_values('Receita', ascending=False).head(10)
```

#**Trabalhando com datas**

```python
#Trasnformando a coluna de data em tipo inteiro
df['Data'] = df['Data'].astype('int64')
```

```python
#Verificando o tipo de dado de cada coluna
df.info()
```

```python
#Transformando coluna de data em data
df['Data'] = pd.to_datetime(df['Data'])
```

```python
df.dtypes
```

```python
#Agrupamento por ano
df.groupby(df['Data'].dt.year)['Receita'].sum()
```

```python
#Criando uma nova coluna com o ano
df['Ano_Venda'] = df['Data'].dt.year
```

```python
df.sample(5)
```

```python
#Extraindo o mês e o dia
df['mes_venda'], df['dia_venda'] = (df['Data'].dt.month, df['Data'].dt.day)
```

```python
df.sample(5)
```

```python
#Retornando a data mais antiga
df['Data'].min()
```

```python
#Retornando a data mais recente
df['Data'].max()
```

```python
#Calculando a diferença de dias
df['diferenca_dias'] = df['Data'] - df['Data'].min()
```

```python
df.sample(5)
```

```python
#Criando a coluna de trimestre
df['trimestre_venda'] = df['Data'].dt.quarter
```

```python
df.sample(5)
```

```python
#Filtrando as vendas de dezembro de 2018
vendas_dezembro_18 = df.loc[(df['Data'].dt.year == 2018) & (df['Data'].dt.month == 12)]
```

```python
vendas_dezembro_18
```

```python
#Filtrando as vendas de janeiro de 2019
vendas_janeiro_19 = df.loc[(df['Data'].dt.year == 2019) & (df['Data'].dt.month == 1)]
```

```python
vendas_janeiro_19.head(10)
```

```python
#Verificando o valor total das vendas em janeiro de 2019
vendas_janeiro_19['Receita'].sum()
```

```python
#Agrupando o valor total das vendas de janeiro de 2019 por cidade
vendas_janeiro_19.groupby('Cidade')['Receita'].sum()
```

```python
#Verificando as dias que mais vendas foram feitas 
df['Data'].value_counts().head(20)
```

```python
df.groupby(df['mes_venda'])['Qtde'].sum().plot(title = 'Total Produtos Vendidos x Mês')
plt.xlabel('Mês')
plt.ylabel('Total Produtos Vendidos')
plt.legend();
```

```python
#Total produtos vendidos por mês
df_2019.groupby(df_2019['mes_venda'])['Qtde'].sum().plot(marker = "x")
plt.xlabel('Mês')
plt.ylabel('Total Produtos Vendidos')
plt.title('Total de produtos vendidos por mês')
plt.legend();
```

```python
#Histograma
plt.hist(df['Qtde'], color="red")
plt.title('Quantidade de produtos vendidos por mês');
```

```python
plt.scatter(x=df_2019['dia_venda'], y = df_2019['Receita'])
plt.title('Receita x Dia da venda');
```

```python
#Salvando em png
df_2019.groupby(df_2019['mes_venda'])['Qtde'].sum().plot(marker = 'o')
plt.title('Quantidade de produtos vendidos x Mês')
plt.xlabel('Mês')
plt.ylabel('Total Produtos Vendidos');
plt.legend()
plt.savefig('grafico QTDE x MES.png')
```
